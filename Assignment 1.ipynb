## 5. Data Visualization

### 1) Univariate analysis of categorical variables

```python
# count plot of column = 'classification'

sns.countplot(x='Classification', data = df)
```

```python
# catplot of column = 'classification'

sns.catplot(x='Classification', data = df)
```

### 2) univariate analysis of numerical columns

```python
# distplot of SL

sns.distplot(df.SL,bins=30, color='g')
```

### 3) Bivariate Analysis: numerical vs Numerical

```python
#lineplot 
# SL and SW

sns.lineplot(x='SL',y='SW', data=df)
```

```python
#lineplot 
# PL and PW

sns.lineplot(x='PL',y='PW',hue='Classification',data=df)
```

```python
#Boxplot of SW

sns.boxplot(x='SW',data=df)
```

```python
# scatter plot of PL vs PW

sns.scatterplot(x='PL',y='PW',hue='Classification',data=df)
```

```python
# scatter plot of SL vs SW

sns.scatterplot(x='SL',y='SW',hue='Classification',data=df)
```

### 4) Bi-variate analysis - numerical vs categorical

```python
sns.barplot(x='SL',y='Classification',data=df)
```

```python
# point plot of SW and Classification
sns.pointplot(x='SW',y='Classification',data=df)
```

### 5) Multivariate Analysis

```python
# Heatmap
sns.heatmap(data=df.corr(),annot=True) 
```
